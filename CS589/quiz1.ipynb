# Problem 1: Top-5 Subsequent words

In this pop quiz, you need to implement a simple algorithm to count the frequency of every word pair in a text corpus, and answer queries such as what are the 5 words that most likely occur after "san" (the answers are "francisco", "diego", "jose", "antonio", "juan"). We will use the text8 corpus from gensim for this quiz.



## Step 0 - Load Dataset


First, we load the dataset from text8.

```python
import gensim.downloader as api
dataset = api.load("text8")
```

## Step 2 - Count the number of word pairs






Implement a Python dictionary which counts the number of every pair in the corpus.

```python
# Initial solution

wordpair2count = {}

### Implement your code below ###

for article in dataset:
    word1 = None
    word2 = None
    for word in article:
        word1 = word2
        word2 = word
        if word1 is None:
            continue
        if not word1 in wordpair2count.keys():
            wordpair2count[word1] = {}
        if not word2 in wordpair2count[word1].keys():
            wordpair2count[word1][word2] = 0
        wordpair2count[word1][word2] += 1
```

```python
# Solution after looking up how to iterate over pairs directly
wordpair2count = {}
for article in dataset:
    for word1, word2 in zip(article, article[1:]):
        if not word1 in wordpair2count.keys():
            wordpair2count[word1] = {}
        if not word2 in wordpair2count[word1].keys():
            wordpair2count[word1][word2] = 0
        wordpair2count[word1][word2] += 1
```

## Step 3 - Answer queries of word frequency

```python
word22count = wordpair2count["san"]

#correct: [('francisco', 576), ('diego', 239), ('jose', 76), ('antonio', 73), ('juan', 53)]
sorted_word22count = sorted(word22count.items(), key = lambda x:x[1], reverse=True)
print(sorted_word22count[0:5])
```

# Problem 2: Regular Expression

Find all the phone numbers in a string: 3 digits followed by 3 digits followed by 4 digits, match the following formats:

xxx xxx xxxx

xxx-xxx-xxxx

(xxx)-xxx-xxxx



```python
### Implement your code below ###
import re

regex_str = r"\(?\d{3}\)?[-\W]\d{3}[-\W]\d{4}"
test_str = "Hello my Number is 201 123 4567 and my friend's number is 201-321-7654 (322)-234-2134"

regex_result = re.findall(r"\(?\d{3}\)?[-\s]\d{3}[-\s]\d{4}", test_str)

print(regex_result)
```
